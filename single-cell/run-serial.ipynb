```python
import os
```

```python
import papermill as pm
```

```python
from sklearn_evaluation import NotebookCollection
```

# Parameters

```python
input_directory = "./inputs/"
output_directory = "./outputs/"

os.makedirs(output_directory, exist_ok=True)

sample_name = "2_DT_2_CD45plus_dense"

experiments = [
    {
        "experiment_id": "candidate-2021-08-11",
        "input_directory": input_directory,
        "output_directory": output_directory,
        "sample_name": sample_name,
        "hvg_n_top_genes": 5000,
        "pca_n_comps": 50,
        "n_neighbors": 30   
    },
    {
        "experiment_id": "test-001",
        "input_directory": input_directory,
        "output_directory": output_directory,
        "sample_name": sample_name,
        "hvg_n_top_genes": 2000,
        "pca_n_comps": 100,
        "n_neighbors": 30   
    },
    {
        "experiment_id": "test-002",
        "input_directory": input_directory,
        "output_directory": output_directory,
        "sample_name": sample_name,
        "hvg_n_top_genes": 1000,
        "pca_n_comps": 100,
        "n_neighbors": 30   
    }    
]
```

# Run Experiments

```python
def run_notebook(param):
        
    path_notebook = os.path.join(
        output_directory, f"{sample_name}.{param['experiment_id']}.ipynb"
    )

    pm.execute_notebook(
        "analysis.template.ipynb",
        output_path=path_notebook,
        parameters=param
    )

    return path_notebook
```

```python
# execute notebooks using papermill

path_notebooks = []

for param in experiments:
    path_notebook = run_notebook(param)
    path_notebooks.append(path_notebook)
```

# Load All Experiment Notebooks

```python
nbs = NotebookCollection(
    paths=path_notebooks,
    ids=[param["experiment_id"] for param in experiments]
)
```
